```python
# Final Database Integration Demo
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Connect to our database
conn_str = "postgresql://postgres:DataScience123!@my-dev-database.cqx0ougy8gof.us-east-1.rds.amazonaws.com:5432/postgres?sslmode=require"
engine = create_engine(conn_str)

# Load our data
df = pd.read_sql("SELECT * FROM sales_data", engine)

print(f"üóÑÔ∏è Database Integration Complete!")
print(f"üìä Loaded {len(df)} records from PostgreSQL")
print(f"üí∞ Total Revenue: ${df['total_amount'].sum():,.2f}")

# Show data
df.head()
```

```python
# Create comprehensive visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Revenue by Product
product_revenue = df.groupby('product')['total_amount'].sum().sort_values(ascending=False)
ax1.bar(product_revenue.index, product_revenue.values, color='skyblue')
ax1.set_title('Revenue by Product')
ax1.set_ylabel('Revenue ($)')
ax1.tick_params(axis='x', rotation=45)

# 2. Orders by Region
region_counts = df['region'].value_counts()
ax2.pie(region_counts.values, labels=region_counts.index, autopct='%1.1f%%')
ax2.set_title('Orders by Region')

# 3. Quantity Distribution
ax3.hist(df['quantity'], bins=5, color='lightgreen', alpha=0.7)
ax3.set_title('Quantity Distribution')
ax3.set_xlabel('Quantity')
ax3.set_ylabel('Frequency')

# 4. Price vs Total Amount
ax4.scatter(df['price'], df['total_amount'], alpha=0.6, color='red')
ax4.set_title('Price vs Total Amount')
ax4.set_xlabel('Unit Price ($)')
ax4.set_ylabel('Total Amount ($)')

plt.tight_layout()
plt.show()

print("‚úÖ Full-Stack Data Science Visualization Complete!")
```
